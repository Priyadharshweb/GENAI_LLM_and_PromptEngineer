```python
from google.colab import userdata
from google.genai import Client

api_key = userdata.get('GENAI')
client = Client(api_key=api_key)

interaction1 = client.interactions.create(
    model='gemini-3.5-flash',
    input="I am Priyadharsini V",
)
print("Response 1:", interaction1.output_text)
```

```python
interaction2 = client.interactions.create(
    model="gemini-3.5-flash",
    input="How many letters in my name?",
    previous_interaction_id=interaction1.id,
)

print("Response 2:", interaction2.output_text)
```

```python
from google.colab import userdata
from google.genai import Client
from google.genai import types

client = Client(api_key=userdata.get("GEN1"))

response = client.models.generate_content(
    model="models/gemini-3.5-flash",
    contents="high",
    config=types.GenerateContentConfig(
        system_instruction="""
You are an assistant that follows this rule:
- If the user says "high", reply with exactly "low".
- Reply with only one word.
""",
        temperature=0.0,
        max_output_tokens=5,
    ),
)

print(response.text)
```

```python
from google.colab import userdata
from google.genai import Client
from google.genai import types

api_key = userdata.get("GEN1")
client = Client(api_key=api_key)

response = client.models.generate_content(
    model="models/gemini-3.5-flash",
    contents="Write a short story about a brave rabbit.",
    config=types.GenerateContentConfig(
        temperature=0.2,
        max_output_tokens=1050,
    ),
)

print(response.text)
```

```python
student_profile = {
    'properties': {
        'totalMarks': {
            'anyOf': [
                {'maximum': 500, 'minimum': 0, 'type': 'integer'},
            ],
            'title': 'Marks',
        },
        'name': {
            'description': "Student's unique name",
            'title': 'name',
            'type': 'string',
        },
    },
    'required': ['name', 'totalMarks'],
    'title': 'Student Schema',
    'type': 'object',
}
```

```python
from google.colab import userdata
from google.genai import Client

api_key = userdata.get('GEN1')
client = Client(api_key=api_key)
interaction3 = client.models.generate_content(
    model='gemini-3.5-flash',
    contents="give me 5 sample standard profile",
    config={
        'response_mime_type': 'application/json',
        'response_json_schema': student_profile
    },
)
```

```python
from groq import Groq
from google.colab import userdata
import json

api_key = userdata.get("groq1")   # Secret name in Colab

client = Groq(api_key=api_key)

completion = client.chat.completions.create(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    messages=[
        {
            "role": "user",
            "content": "Generate marks for 5 students in the given JSON schema."
        }
    ],
    temperature=1,
    max_completion_tokens=1024,
    response_format={
        "type": "json_schema",
        "json_schema": {
            "name": "student_profile",
            "schema": student_profile
        }
    }
)

result = json.loads(completion.choices[0].message.content)
print(result)
```
