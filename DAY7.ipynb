```python
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
from datetime import datetime

@tool
def get_current_time():
    """Return the current date and time
    """
    return datetime.datetime.now().isoformat()

@tool
def add(a,b):
  """adds two integers and return the output"""
  print("ADD:"+a+b)
  return a+b

api_key = userdata.get('GEN2')
model = ChatGoogleGenerativeAI(
    api_key=api_key,
    model="gemini-3.6-flash",
)
model=model.bind_tools([get_current_time])
response3 = model.invoke("what is current date and time")
print(response3.content)
```

```python
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
from datetime import datetime

@tool
def get_current_time():
    """Return the current date and time
    """
    return datetime.datetime.now().isoformat()

def add(a,b):
  """adds two integers and return the output"""
  # Corrected print statement to handle integer concatenation
  print(f"ADD:{a+b}")
  return a+b

api_key = userdata.get('GEN2')
model = ChatGoogleGenerativeAI(
    api_key=api_key,
    model="gemini-3.6-flash",
)
model=model.bind_tools([get_current_time])
response3 = model.invoke("what is current date and time")
print(response3.content)
num1 = 5
num2 = 7
# Now 'add' can be called directly as it's no longer a StructuredTool
result = add(num1, num2)
print(f"The sum of {num1} and {num2} is: {result}")
```

```python
import os
from langchain_core.messages import HumanMessage
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_google_genai import ChatGoogleGenerativeAI

model=ChatGoogleGenerativeAI(
    api_key=api_key,
    model="gemini-3.6-flash"
)
store={}
def get_session_history(session_id:str):
  if session_id not in store:
    store[session_id]=InMemoryChatMessageHistory
  return store[session_id]

model=RunnableWithMessageHistory(
    model,
    get_session_history
)
config = {"configurable": {"session_id": "session_alpha"}}
```

```python
from langchain_core.messages import HumanMessage
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_google_genai import ChatGoogleGenerativeAI
base_model = ChatGoogleGenerativeAI(
    api_key=api_key,
    model="gemini-3.6-flash"
)

local_store = {}
def get_local_session_history(session_id:str):
  if session_id not in local_store:

    local_store[session_id]=InMemoryChatMessageHistory()
  return local_store[session_id]
model = RunnableWithMessageHistory(
    base_model,
    get_local_session_history
)

res1 = model.invoke(
   [HumanMessage(content="My name is Priyadharsini")],
    config=config
)

print(res1.content)
```
