```python
import base64
from google.colab import userdata
from google.genai import Client

api_key = userdata.get('GEN1')
client = Client(api_key=api_key)
with open("/content/licensed-image.webp", "rb") as f:
      image_bytes = f.read()
```

```python
image_bytes
```

```python
image_b64 = base64.b64encode(image_bytes).decode("utf-8")
image_b64
```

```python
interaction = client.interactions.create(
    model="gemini-3.5-flash",
    input=[
        {
            "type":"image",
            "data":image_b64,
            "mime_type":"image/jpeg"
        },
         {
            "type": "text",
            "text":"define the image in 2 line"
        }
    ]
)
```

```python
interaction.output_text
```

```python
interaction
```

```python
interaction2 = client.interactions.create(
    model="gemini-3.5-flash",
    input="explain his century",
    # continue to that chat
    previous_interaction_id=interaction.id
)
```

```python
interaction2.output_text
```

```python
interaction2
```

```python
interaction3 = client.interactions.create(
    model="gemini-3.5-flash",
    input= "explain about his history in 2 lines",
    previous_interaction_id=interaction2.id
)
```

```python
import time
uploaded_video = client.files.upload(file="/content/Recording 2025-12-03 172801.mp4")

file_status = client.files.get(name=uploaded_video.name)
while file_status.state.name == "PROCESSING":
    print(".", end="", flush=True)
    time.sleep(5)  # Wait 5 seconds before checking again
    file_status = client.files.get(name=uploaded_video.name)

if file_status.state.name == "FAILED":
    raise ValueError(f"Video processing failed: {file_status.error}")

interaction4 = client.interactions.create(
    model="gemini-3.5-flash",
    input=[
        {
            "type": "image",
            "data": image_b64,
            "mime_type": "image/jpeg"
        },
        {
            "type": "video",
            "uri": uploaded_video.uri,
            "mime_type": "video/mp4"
        },
        {
            "type": "text",
            "text": "Summarize this video. and explain few words about the image also"
        }

    ]
)
print(interaction4.output_text)
```
